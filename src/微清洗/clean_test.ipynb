```python
import pandas as pd

def compare_dataframe_features(df1, df2):
    """
    对比两个 DataFrame 的特征（列名），返回共同特征、df1 特有特征和 df2 特有特征。

    参数：
        df1: 第一个 DataFrame。
        df2: 第二个 DataFrame。

    返回值：
        一个包含三个列表的元组：
        - 共同特征列表
        - df1 特有特征列表
        - df2 特有特征列表
    """

    # 将 DataFrame 的列名转换为集合
    df1_features = set(df1.columns)
    df2_features = set(df2.columns)

    # 计算共同特征（交集）
    common_features = list(df1_features.intersection(df2_features))

    # 计算 df1 特有特征（差集）
    df1_unique_features = list(df1_features.difference(df2_features))

    # 计算 df2 特有特征（差集）
    df2_unique_features = list(df2_features.difference(df1_features))

    return common_features, df1_unique_features, df2_unique_features

common, mimic_unique, eicu_unique = compare_dataframe_features(mimic_deficiency, eicu_deficiency)

# 打印结果
print("共同特征:", common)
print("mimic_data 特有特征:", mimic_unique)
print("eicu_data 特有特征:", eicu_unique)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor

def check_collinearity(X):
    """
    全面检查 DataFrame 中特征的共线性情况，包括相关性热图和 VIF 值。

    参数:
        X (pd.DataFrame): 输入的特征 DataFrame。

    返回:
        None: 函数会打印相关性热图和 VIF 值，并输出文字描述。
    """

    print("==================== 共线性检查 ====================\n")

    # 1. 相关性分析 - 相关系数矩阵和热图
    print("1. 特征相关性分析 (Correlation Analysis):\n")
    correlation_matrix = X.corr(method='pearson') # 计算 Pearson 相关系数
    plt.figure(figsize=(15, 12)) # 调整图像大小以适应特征数量
    sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f") # annot=False 避免在单元格中显示数值，如果特征太多可以省略
    plt.title("Correlation Heatmap")
    plt.show() # 显示热图

    # 2. 方差膨胀因子 (VIF)
    print("\n2. 方差膨胀因子 (Variance Inflation Factor - VIF):\n")
    vif_values = pd.DataFrame()
    vif_values["Feature"] = X.columns
    vif_values["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    print(vif_values.sort_values(by='VIF', ascending=False).to_string()) # 打印 VIF 值，并按降序排列

    print("\n==================== 共线性检查完成 ====================\n")


check_collinearity(mimic_deficiency) # 对 X_train 进行共线性检查
```
